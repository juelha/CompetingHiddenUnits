```python
# basics 
import numpy as np
import matplotlib.pyplot as plt
import os
from sklearn import datasets

from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import LearningCurveDisplay, ShuffleSplit
from sklearn.tree import DecisionTreeClassifier

from tqdm import tqdm 
#import pandas as pd

import visu
```

```python

file_name = 'xor' + '.csv'


# get save path 
relative_path =  '/../data'
save_path = os.getcwd() + relative_path # https://stackoverflow.com/questions/39125532/file-does-not-exist-in-jupyter-notebook
full_path = os.path.join(save_path, file_name)
assert  os.path.exists(full_path), f'Path {relative_path}/{file_name} does not exist'
data = pd.read_csv(full_path)

labels = data.pop(data.columns[-1]) # always use last column


# split 
x_train, x_test = np.split(data,[int(0.8 * len(data))])
y_train, y_test = np.split(labels,[int(0.8 * len(labels))])  

clf = DecisionTreeClassifier(random_state=0)
clf.fit(x_train, y_train)

pred_test = clf.predict(x_test)

# #print('Accuracy: {:.2f}'.format(accuracy_score(y_test, pred_test)))




LearningCurveDisplay.from_estimator(clf, x_train, y_train, scoring='accuracy')

plt.show()
```
